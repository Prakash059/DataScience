# Getting the content of  the website using webscraping

```python
import requests
```

```python
# Importing BeautifulSoup4 library
import bs4
```

```python
res = requests.get('https://en.wikipedia.org/wiki/Sachin_Tendulkar')
```

```python
res
```

```python
type(res)
```

```python
# All content of website
res.text
```

```python
# Convert res object into BeautifulSoup object
soup = bs4.BeautifulSoup(res.text,'lxml')
```

```python
type(soup)
```

```python
soup
```

```python
title = soup.select('title')
```

```python
title
```

```python
type(title)
```

```python
title[0].getText()
```

```python
div = soup.select('div')
```

```python
div
```

```python
# For class selection use '.' before the class name
clas = soup.select('.toc')
```

```python
clas
```

```python
for i in soup.select('.toc'):
    print(i.text)
```

```python
for link in soup.findAll('a',href=True):
    print(link['href'])
```
