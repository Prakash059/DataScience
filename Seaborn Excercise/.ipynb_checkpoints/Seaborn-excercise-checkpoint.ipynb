```python
import seaborn as sns
```

```python
myTitanic = sns.load_dataset('titanic')
```

```python
myTitanic.head()
```

```python
sns.distplot(myTitanic['fare'])
```

```python
sns.distplot(myTitanic['fare'],kde=False)
```

```python
sns.jointplot(x='age',y='fare',data = myTitanic)
```

```python
sns.jointplot(x='age',y='fare',data = myTitanic,kind='hex')
```

```python
sns.jointplot(x='age',y='fare',data = myTitanic,kind='reg')
```

#Factor plot and fat consumption data

```python
%matplotlib inline
#use of the above line-no need to write plt.show()
```

```python
myDf = sns.load_dataset('exercise')
```

```python
myDf.head()
```

```python
myDf.count()
```

```python
sns.factorplot(x='time',y='pulse',hue='kind',data = myDf)
```

```python
sns.factorplot(x='time',y='pulse',hue='kind',data = myDf,col = 'diet')
```
